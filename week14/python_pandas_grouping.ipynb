### Python- Pandas DataFrame- Grouping and Mapping

```python
import numpy as np

from pandas import Series,DataFrame
import pandas as pd
```

```python
df = DataFrame({'city':['Alma', 'Brian Head', 'Fox Park', 'Lake May'],
               'altitude':[10355,9800, 9062, 8966]})
               
df
```

```python
st_map={'Alma':'Colorado', 'Brian Head': 'Utah', 'Fox Park':'Wyoming', 'Lake Mary':'California'}
st_map
```

```python
df['state'] = df['city'].map(st_map)
df
```

```python
def meter(x):
    return x / 3.2808
```

```python
df['altitude'] = df['altitude'].apply(meter)

df
```

```python
df =  DataFrame(np.random.randn(4,3), columns=list('bde'),
               index= ['UT', 'OH', 'TX', 'OR'])
df
```

```python
def dif(x):
    return x.max() - x.min()
```

```python
df.apply(dif)
```

```python
df.apply(dif, axis=1)
```

```python
df.apply(pd.value_counts).fillna(0)
```

```python
df = DataFrame({'k1': ['a', 'a', 'b', 'b', 'c'],
               'k2': ['one', 'two', 'one', 'two', 'one' ],
               'dataset1': np.random.randn(5),
               'dataset2': np.random.randn(5)})
df
```

```python
df.groupby('k1').count()
```

```python
df.groupby('k1').size()
```

```python
df.groupby(['k1', 'k2']).count()
```
