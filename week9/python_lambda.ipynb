### Python - lambda functions

```python
def double(num):
    return num * 2
print(double(3))
```

```python
def square(x):
    return x**2

square(5)
```

```python
double = lambda x:x*2

print(double(4))
```

```python
# find larger function using lambda function

larger = lambda a,b: a if a > b else b
    
print(larger(5,10))
```

```python
nums = [1,2,3,4,5,6,7]

map(square, nums)
```

```python
mylist = list(map(square, nums))
```

```python
mylist
```

```python
for item in map(square,nums):
    print(item)
```

```python
def evenwords(w):
        if (len(w)%2 == 0):
            return "EVEN"
        else:
            return w[0:2]
evenwords("Dodge")
    
```

```python
words = ["ford", "nissan" , "GMC", "Mazda" , "Dodge", "BMW" ]

list(map(evenwords, words))
```

```python
list(map(lambda x:x**2, nums))
```

```python
list(map(lambda w:w[0], words))
```

```python
def even(num):
    return num%2 == 0

even(5)
```
